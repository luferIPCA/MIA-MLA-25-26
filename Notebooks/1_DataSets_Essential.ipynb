## 2 - Python Object Oriented (POO)

Object programing means the existence of classes (models) and objects (instances)


### Classes definition

````python
#Class definition
#lufer
class Person:
  #-----
  #Attributes | state
  name=""
  age=1
  #-----
  #Methods | Behaviour
  # assign values to object properties, or other operations that are necessary
  # to do when the object is being created
  #Constructor
  def __init__(self, name, age):
    #print("Inside __init__")
    self.name = name
    self.age = age
  #-----
  #  controls what should be returned when the class object is represented as a string
  def __str__(self):
    return f"Name: {self.name} (Age: {self.age})"
  #-----
  # other methods
  def myName(self):
    print("Hello my name is  " + self.name)

  #class methods
  def myAge(self):
    print("I am " + str(self.age) + " years old")
````

### Objects creation and using

````python
#Objects
p1 = Person("Ana Gustavo","12")   #uses __init__

print(p1)                         #uses __str__
print(p1.age)
print (f"Age: ({p1.age}) - Name: ({p1.name}) " )
p1.myName();
p1.name
````

### Python's .*\_\_call__()* Method [link text](https://)

*\_\_call()__* is an alternative to make "callable" methods or variables. Many actual python versions does not support this.

**NOTE:**
* Every Python class is callable, meaning that calling it creates a new instance of the class
* *\_\_call__* can generate complexity

````python
#callable(abs)
#check if it is callable
callable (Person.myAge)
````

````python
#Another  example
class PersonII:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def Hi(self):
        return f"Hi, I'm {self.name} and I'm {self.age} years old."

    def MyAge(self):
        return f"My age is {self.age}"

    #def __call__(self):
    #    # Calling the object will call MyAge()
    #    return self.MyAge()

    def __call__(self, action=None):
        if action is None:
            # Default callable behavior
            return self.MyAge()

        if action == "Hi":
            return self.Hi()

        return f"Unknown action: {action}"
````

````python
#using it

a=PersonII("ola",12)
a.Hi()
````

````python
#using the callable method
a()

a("Hi")
````

### Basic Statistics over a Dataset

The basic statisics of a dataset can be obtained using the python function *describe()*.

1. For Numeric Columns (default behavior), it provides summary statistics such as:
 * Count: The number of non-missing values.
 * Mean: The average of the values.
 * Std: The standard deviation (measure of spread).
 * Min: The smallest value.
 * 25%: The first quartile (25th percentile).
 * 50%: The median (50th percentile).
 * 75%: The third quartile (75th percentile).
 * Max: The largest value.
2.  For Non-Numeric Columns (if specified using include='all' or include=object), it provides:
 * Count: The number of non-missing values.
 * Unique: The number of unique values.
 * Top: The most frequent value.
 * Freq: The frequency of the top value.

3. **Key Notes about 'describe()'**
* Non-numeric data is ignored unless explicitly included using include.
* The method helps quickly identify:
 * Missing data (count vs. total rows).
 * Outliers (e.g., large differences between min, max, and mean).
 * Skewness (if the mean and median differ significantly).

 * **Default Percentiles**: percentiles=None computes the 25th,  50th, and 75th percentiles by default.
* **Default Inclusion**: exclude=None includes numeric columns only, unless explicitly overridden.
* Explicit Parameters:
 * Use percentiles to customize which percentiles to compute.
 * Use include='all' to analyze both numeric and non-numeric columns.
 * Use exclude to omit specific data types (e.g., exclude=[float]).
