### Basic Statistics over a Dataset

Basic statistics with 'describe()'

1. For Numeric Columns (default behavior):

* It provides summary statistics such as:
 * Count: The number of non-missing values.
 * Mean: The average of the values.
 * Std: The standard deviation (measure of spread).
 * Min: The smallest value.
 * 25%: The first quartile (25th percentile).
 * 50%: The median (50th percentile).
 * 75%: The third quartile (75th percentile).
 * Max: The largest value.
2.  For Non-Numeric Columns (if specified using include='all' or include=object):

* It provides:
 * Count: The number of non-missing values.
 * Unique: The number of unique values.
 * Top: The most frequent value.
 * Freq: The frequency of the top value.

3. **Key Notes about 'describe()'**
* Non-numeric data is ignored unless explicitly included using include.
* The method helps quickly identify:
 * Missing data (count vs. total rows).
 * Outliers (e.g., large differences between min, max, and mean).
 * Skewness (if the mean and median differ significantly).

 * **Default Percentiles**: percentiles=None computes the 25th,  50th, and 75th percentiles by default.
* **Default Inclusion**: exclude=None includes numeric columns only, unless explicitly overridden.
* Explicit Parameters:
 * Use percentiles to customize which percentiles to compute.
 * Use include='all' to analyze both numeric and non-numeric columns.
 * Use exclude to omit specific data types (e.g., exclude=[float]).

```
# What does this function?
import random

def generateNewValues(existing_values, number_new_values):
    base_values = ["V1", "V2", "V3", "V4", "V5"]
    prefix = ["A", "B", "C", "D"]
    new_values = set()

    while len(new_values) < number_new_values:
        # Combine a prefix with a base value
        v = f"{random.choice(prefix)} {random.choice(base_values)}"

        # Ensure the new value is not in the existing list
        if v not in existing_values:
            new_values.add(v)

    return list(new_values)

# Step 1: Extract existing colors
existing_values = ds['Colour'].dropna().unique().tolist()
#check it
#existing_values

#test it
new_values = generateNewValues(existing_values,10)
new_values

#-------------------------------------------------------------------------------------------------------
#Goal: Combines the existing values with the autonomously generated ones to form a more complete list.

#Try to explore this function to generate new colours with a classification prefix, like "Brigh Red"
# Suggestion:
# prefix = ["Bright", "Deep", "Soft", "Dark", "Light"]
# base_values = ["Red", "Blue", "Green", "Yellow", "Purple", "Orange", "Pink", "Cyan"]
```

**And this new example?**

```
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Step 1: Example dataset with numerical and categorical features
data = pd.DataFrame({
    'age': [23, 45, 34, 50, 29],
    'salary': [50000, 70000, 60000, 80000, 55000],
    'gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'department': ['HR', 'Engineering', 'HR', 'Engineering', 'Marketing']
})

# Step 2: Generate synthetic data for numerical columns
# Calculate means and stds for numerical columns
numerical_cols = ['age', 'salary']
means = data[numerical_cols].mean()
stds = data[numerical_cols].std()

# Number of synthetic samples to generate
qty_synth = 10

# Generate synthetic data for numerical features based on normal distribution
synthetic_numerical_data = np.random.normal(loc=means.values, scale=stds.values, size=(qty_synth, len(numerical_cols)))

# Convert synthetic numerical data into a DataFrame
synthetic_numerical_df = pd.DataFrame(synthetic_numerical_data, columns=numerical_cols)

# Step 3: Generate synthetic data for categorical columns
# For each categorical column, we calculate the frequencies (or median)
categorical_cols = ['gender', 'department']
synthetic_categorical_data = {}

for col in categorical_cols:
    # Get the frequency distribution for each categorical column
    category_frequencies = data[col].value_counts(normalize=True)

    # Sample synthetic data based on the frequencies
    synthetic_categorical_data[col] = np.random.choice(category_frequencies.index,
                                                       size=qty_synth,
                                                       p=category_frequencies.values) #p = probability=frequency

# Step 4: Combine the numerical and categorical synthetic data
synthetic_data = pd.DataFrame(synthetic_categorical_data)
synthetic_data = pd.concat([synthetic_numerical_df, synthetic_data], axis=1)

# Step 5: Display the synthetic dataset
print(synthetic_data)
```
